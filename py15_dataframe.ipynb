## 카테고리 별 2개 이상 변수(컬럼)의 통계값 찾기

```python
# 성별 영수증, 팁의 평균
tips.groupby('sex')[['total_bill', 'tip']].mean()
```

```python
tips.groupby('smoker')[['total_bill', 'tip']].max()
```

```python
# 성별, 흡연여부별 영수증과, 팁의 평균
tips.groupby(['sex', 'smoker'])[['total_bill', 'tip']].mean()
```

## 그룹별 2개 이상의 통계 함수를 적용하는 방법.

## aggregate(), agg() 함수

```python
# 성별 팁의 평균 최댓값, 중앙값(중위값), 최솟값
tips.groupby('sex').tip.agg([np.mean, np.min, np.median, np.max]) # aggregate
```

```python
# 성별, 흡연여부별 영수증과 팁의 평균과, 표준편차를 출력.
tips.groupby(['sex', 'smoker'])[['total_bill', 'tip']].agg([np.mean, np.var, np.std])
```

# `pivot_table` 함수(메서드)

* Parameters:
  * values: 통계 함수를 적용할 변수(컬럼)
  * index: 결과 데이터프레임에서 행 인덱스로 사용할 변수(컬럼)
  * columns: 결과 데이터프레임에서 열 인덱스로 사용할 변수(컬럼)
  * aggfunc: aggregation function. 기본값은 평균(mean)

```python
# 성별 팁의 평균
# pivot_table()의 aggfunc의 기본값은 aggfunc=mean
tips.pivot_table(values='tip', index='sex')
```

```python
# 성별 흡연여부별 팁의 평균.
tips.pivot_table(values='tip', index=['sex', 'smoker'])
```

```python
# 성별, 흡연여부별 팁의 평균
tips.pivot_table(values='tip', index='sex', columns='smoker')
```

```python
# 성별, 흡연여부별, 요일별, 시간별 팁의 평균
tips.pivot_table(values='tip', index=['sex', 'smoker'], columns=['day', 'time'])
```

```python
# 요일별 시간별 영수증과 팁의 평균, 표준편차 출력
tips.pivot_table(values=['total_bill' ,'tip'], index=['day', 'time'], aggfunc=[np.mean, np.std])
```
