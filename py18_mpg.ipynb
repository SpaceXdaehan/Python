## 2x2 subplot을 생성. scatter plot
  * mpg ~ cylinders
  * mpg ~ horsepower
  * mpg ~ weight
  * mpg ~ acceleration

```python
variables = ['cylinders', 'horsepower', 'weight', 'acceleration'] # 독립변수

# 2x2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

# 각각의 subplot에 산점도 그리기:
for i in range(2):
  for j in range(2):
   idx = 2 * i + j
   sns.scatterplot(data=mpg, x=variables[idx], y='mpg', ax=ax[i, j])
plt.show()
```

## 상관 관계, 상관 계수

```python
sns.pairplot(data=mpg) # 숫자 타입 변수들 간의 상관 관계를 파악
plt.show()
```

```python
sns.pairplot(data=mpg, hue='origin')
plt.show()
```

# 상관 계수 (correlation coefficient):

* 두 변수의 상관 관계를 수치로 표현.
* -1 <= 상관 계수 <= 1
* 상관 계수가 1 또는 -1에 가까울 수록 상관 관계가 크다고 하고, 0에 가까울 수록 상관 관계가 작다고 말한다.
  * 상관 계수가 양수인 경우 양의 상관 관계가 있다고 말함.
  * 상관 계수가 음수인 경우 음의 상관 관계가 있다고 말함.
* `pd.DataFrame.corr()` 메서드는 상관 계수들을 계산하는데, 기본 방식은 pearson 상관 계수를 계산해줌.

```python
mpg.corr(numeric_only=True) # umeric_only = 숫자 타입들만 상관 계수를 계산하겠다.
```
