## indexing, slicing(Series의 두가지 인덱스)

* `iloc`: integer location. 정수 인덱스(ndarray(넘파이배열의)의 기본 인덱스) 기반으로 원소 참조 또는 슬라이싱.
* `loc`: location. 레이블 기반의 원소 참조 또는 슬라이싱.

```python
np.random.seed(1)
s1 = pd.Series(data=np.random.randint(-10, 10, size=6))
s1
```

```python
# iloc를 사용한 인덱싱
print(s1.iloc[0]) # Series에서 첫번째 원소
print(s1.iloc[-1]) # Series에서 마지막 원소
```

```python
# iloc를 사용한 슬라이싱
print(s1.iloc[:3]) # Series에서 첫 3개 원소를 자르기
print(s1.iloc[-3:]) # Series에서 마지막 원소 3개 자르기
```

```python
# loc를 사용한 인덱싱
print(s1.loc[0]) # Series의 첫번째 원소
print(s1.loc[5]) # Series의 마지막 원소 - 음수 인덱스를 사용할 수 없음.
```

```python
# loc을 사용한 슬라이싱
print(s1.loc[0:2])
print(s1.loc[3:5])
```

# DataFrame

* 데이터를 2차원(행 row, 열 column) 형태로 저장하는 데이터 타입(클래스).
* 데이터베이스의 테이블과 비슷한 구조.
* DataFram에서 컬럼은 Series 타입이다.

```python
# dict
data = {
    'city': ['seoul'] * 3 + ['goyang'] * 3,
    'year': [2021, 2022, 2023] * 2,
    'pop': np.random.random(size=6)
}
data
```

```python
df = pd.DataFrame(data)
```

## DataFrame에서 컬럼 선택 두가지 방법

* `df['column_name']`

  * 컬럼 이름이 DataFrame 객체가 원래 가지고 있는 속성(필드, 메서드) 이름과 같은 경우에는 사용할 수 없음.

* boolean indexing을 사용할 때에는 파이썬의 논리 연산자`(and, or, not)`를 사용할 수 없음!
* boolean indexing에서는 `&, |, ~` 연산자를 사용함!
* 조건식과 조건식은 ()를 사용해서 구분, 연산의 순서를 명시해야함.

```python
# city가 seoul이거나, pop이 0.65 이하인 행 들을 선택:
df[(df['city'] == 'seoul') | (df['pop'] <= 0.65)]
```

```python
# 고양의 2023년 자료를 선택
df[(df.city == 'goyang') | (df.year == 2023)]
```

```python
means = []
for x in range(1, 6):
  math_mean = exam[exam['class'] == x].math.mean()
  means.append(math_mean)

print(means)

exam_dict = {
    'class': np.arange(1, 6),
    'math_mean': means
}
print(exam_dict)

df = pd.DataFrame(exam_dict)
df
```
