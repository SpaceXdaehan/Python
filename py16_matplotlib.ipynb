# `matplotlib.pyplot` 모듈을 사용한 시각화

1. `figure` 객체 생성
1. `figure` 객체에 subplot(axis)을 생성해서 추가
1. subplot에 그래프 그림
1. 그래프에 여러가지 옵션들을 설정해줌
1. 그래프를 보여주기

```python
# 1. figure 객체 생성
figure = plt.figure()

# 2. 생성된 figure 객체에 subplot(axis)을 추가
# subplot은 index가 1부터 시작함.
# axis = figure.add_subplot(1, 1, 1) # (row, column, index)
axis = figure.add_subplot(111) # 위랑 같은 코드

# 3. subplot에 그래프 종류를 선택해서 그림.
axis.hist(x=tips['total_bill'])

# 4. 그래프에 여러가지 옵션을 설정한다.
axis.set_xlabel('Total Bill') # x축
axis.set_ylabel('Count') # y 축
axis.set_title('Histogram') # 상단 title
axis.grid() # figure 칸나눔

#5 그래프 보여주기
plt.show()
```

```python
# 1. figure를 생성해줌
fig = plt.figure(figsize=(6, 8))

# 2. subplot을 figure에 추가
ax1 = fig.add_subplot(2, 1, 1) # add_subplot(211)
ax2 = fig.add_subplot(2, 1, 2) # add_subplot(212)

# 3. 그래프 그리기
ax1.hist(x=tips.total_bill)
ax2.scatter(x=tips.total_bill, y=tips.tip)

# 4. 옵션
ax1.grid()
ax2.grid()

# 5. 그래프 보여주기
plt.show()
```

figure에 두개의 subplot을 상하로 배치하고 그래프 그리기.

figure에 2개의 subplot을 좌우로 배치하고 hist, scatter 그래프를 그리기

```python
fig = plt.figure(figsize=(12, 4))

ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

ax1.hist(x=tips.total_bill, color='violet')
ax2.scatter(x=tips.total_bill, y=tips.tip, color='limegreen')

ax1.grid()
ax1.set_xlabel('Total Bill') # x축
ax1.set_ylabel('Count')
ax1.set_title('Histogram')

ax2.grid()
ax2.set_xlabel('Total Bill')
ax2.set_ylabel('Count')
ax2.set_title('Scatter')

plt.show()
```

# `pyplot` 모듈의 top-level 함수

figure 생성 + subplot 1개 생성 + 그래프 종류

hist 함수 파라미터:

* x: x축에 사용할 변수 이름. 히스토그램을 그리기 위한 자료.
* bins: 막대의 개수, 나누는 구간
* color: 막대 안쪽을 채우는 색깔
* edgecolor: 막대 테두리 색깔
* density: False인 경우 = 개수, True인 경우 = 밀도(비율)

## Histogram

* 연속형 변수를 구간으로 나눠서, 그 구간 안에 포함된 자료의 개수를 막대의 높이로 표현한 그래프.

* 연속형 변수의 분포를 알 수 있음.

```python
# 1. figure를 생성하고 2. subplot 1개를 생성하고 3. histogram을 그린다.
plt.hist(x=tips.total_bill, color='orange', edgecolor='black', bins=10, density='True')

# 4. 옵션을 설정한다.
plt.xlabel('Total bill')
plt.ylabel('Count')
plt.title('Histogram')
plt.grid()

# 5. 그래프를 보여준다.
plt.show()
```

## 산점도(Scatter plot)

두 변수(컬럼) 사이의 상관 관계를 파악하기 위해서 사용.

```python
# tip ~ total_bill 사이의 상관 관계 그래프
plt.scatter(x=tips.total_bill, y=tips.tip,
            alpha=0.3, # alpha: 0(투명) ~ 1.0(불투명)의 값을 줄 수 있음
            s=30, # s: size. 마커(점)의 크기.
            marker='*', # marker: 마커의 모양(종류). ('o', 's', '^', '*', 'v')
            color='gold')
plt.title('Scatter Chart')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.grid()

plt.show()
```

```python
# tip ~ size 상관 관계
plt.scatter(x=tips['size'], y=tips.tip, color='red', alpha=0.3, s=30, marker='^')
plt.title('Scatter Chart')
plt.xlabel('Size')
plt.ylabel('Tip')
plt.grid()

plt.show()
```

## 선 그래프(Line plot)

주로 시계열 데이터(시간에 따라서 변하는 데이터: 주가, 환율, 날씨, ...)를 시각화.

```python
np.random.seed(1) # 난수를 생성하는 순서를 설정.
df = pd.DataFrame(data={
    'time': np.arange(1, 11), # [1, 2, 3, ..., 9, 10]
    'price': np.random.rand(10) # [0, 1.0)범위의 난수 10개를 갖는 1차원 배열.
})
df
```

```python
# price ~ time
plt.plot(df.time, df.price, 'k*:')
# fomat 문자열(3번째 arg): '색깔/마커모양/선의종류'
# 색깔: b(lue), r(ed), g(reen), blac(k), ...
# 마커 종류: o, ^, v, *, s, +
# 선의 종류: -(solid line), :(dotted line), --(dashed line)

plt.title('Line Plot')
plt.xlabel('Time')
plt.ylabel('Price')
plt.grid()

plt.show
```

```python
# variable-length keyword argument(**) 사용법

plt.plot(df.time, df.price,
         c='crimson',
         marker='*',
         linestyle='dotted')
plt.grid()

plt.show()
```

### 1개의 subplot에 선 그래프 3개를 중첩시켜서 표현하기.

```python
np.random.seed(1)
df = pd.DataFrame(data={
    'time': np.arange(1, 11),
    'price1': np.random.rand(10),
    'price2': np.random.rand(10),
    'price3': np.random.rand(10)
})
df
```

```python
# price1 ~ time
plt.plot(df.time, df.price1, label='price1', marker='*', linestyle='dashed')
plt.plot(df.time, df.price2, label='price2', marker='^', linestyle='dotted')
plt.plot(df.time, df.price3, label='price3', marker='s', linestyle='dashed')

plt.legend() # 범례(legend)를 화면에 표시
plt.grid()

plt.show()
```

## Ex 3.
* tip ~ total_bill의 상관관계를 scatter plot으로 그리세요.
* 요일별로 점의 색깔, 모양을 다르게 표현하세요.
범례를 표시.

```python
days = tips.day.unique()
markers = ['o', '^', 's', '*']
colors = ['red', 'orange', 'pink', 'deepskyblue']

for d, m, c in zip(days, markers, colors):
  subset = tips[tips.day == d]
  plt.scatter(x=subset.total_bill, y=subset.tip, c=c, marker=m, alpha=0.35, label=d)

plt.legend()
plt.grid()
plt.xlabel('Total bill')
plt.ylabel('Tip')
plt.show()
```
