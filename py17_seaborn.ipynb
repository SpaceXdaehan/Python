```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

# Box plot

```python
# tip 컬럼의 box plot
sns.boxplot(data=tips, y='tip')
plt.show()
```

```python
sns.boxplot(data=tips, x='tip')
plt.show()
```

## 그룹(카테고리)별 box plot

```python
# 성별 tip의 box plot
sns.boxplot(data=tips, x='sex', y='tip')
plt.show()
```

```python
sns.boxplot(data=tips, x='tip', y='sex')
plt.show()
```

## 요일별 tip의 box plot

```python
sns.boxplot(data=tips, x='day', y='tip')
plt.show()
```

```python
sns.boxplot(data=tips, x='tip', y='day')
plt.show()
```

* subplot 2개를 생성
* 각각의 subplot에 흡연여부별 tips의 box plot, 시간별 box plot를 그리세요.

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 6))
sns.boxplot(data=tips, x='smoker', y='tip', ax=ax[0])
sns.boxplot(data=tips, x='time', y='tip', ax=ax[1])

plt.show()
```

## 막대 그래프

## `countplot()`함수

카테고리 타입 변수의 빈도수를 막대로 표현한 그래프.

```python
# 성별 빈도수 세로 막대 그래프를 countplot을 이용해서 출력
sns.countplot(data=tips, x='sex')
plt.show()
```

```python
# 성별 빈도수를 가로 막대 그래프 countplot를 사용해서 출력
sns.countplot(data=tips, y='sex')
plt.show()
```

```python
# 요일 빈도수 막대 그래프
sns.countplot(data=tips, x='day')
plt.show()
```

```python
sns.countplot(data=tips, y='day')
plt.show()
```

## hue: 파라미터 사용

* 개수가 제한되어 있는 카테고리만 사용할 수 있음.(연속되는 변수를 사용할 수 없음)

```python
# 요일별 성별 빈도수
tips.groupby(['day', 'sex']).tip.count()
```

```python
# hue 파라미터: 그룹(카테고리) 별로 다른 색깔을 설정할 때.
sns.countplot(data=tips, x='day', hue='sex')
plt.show()
```

```python
sns.countplot(data=tips, x='sex', hue='day')
plt.show()
```

```python
# 성별 흡연여부별 빈도수
tips.groupby(['sex', 'smoker']).tip.count()
```

```python
sns.countplot(data=tips, x='sex', hue='smoker')
plt.show()
```

```python
sns.countplot(data=tips, x='smoker', hue='sex')
plt.show()
```

## `barplot()` 함수

* 그룹(카테고리) 별로 변수의 통계량(평균, 표준편차, 중앙값, ...)을 막대 그래프로 표현하는 함수.
* 통계량 계산 함수의 기본값은 평균(mean).
* 카테고리 타입 변수와 숫자 타입 변수를 함께 시각화 한다.

```python
# 성별 tip의 평균
tips.groupby('sex').tip.mean()
```

estimator='mean': 기본값, 평균 계산

```python
sns.barplot(data=tips, x='sex', y='tip') # estimator='mean': 기본값, 평균 계산
plt.show()
```

```python
tips.groupby('sex').tip.median()
```

```python
sns.barplot(data=tips, x='sex', y='tip', estimator='median')
plt.show()
```

```python
# 흡연여부별 tip의 평균을 시각화
tips.groupby(['smoker', 'tip']).tip.mean()
```

```python
sns.barplot(data=tips, x='smoker', y='tip')
plt.show()
```

```python
# 성별 흡연여부를 시각화
tips.groupby(['sex', 'smoker']).tip.mean()
```

```python
sns.barplot(data=tips, x='sex', y='tip', hue='smoker')
plt.show()
```

# 산점도: `scatterplot()`

* tip ~ total_bill
* 성별에 따라서 점의 색깔이 다르게 표현

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
```

* tip ~ total_bill
* 성별에 따라서 점의 색깔을 다르게.
* 흡연여부에 따라서 점의 모양을 다르게.

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='smoker')
plt.show()
```

* tip ~ total_bill
* 점의 색깔은 성별에 따라서 다르게 표현
* 점의 모양은 흡연여부에 따라서 다르게 표현
* 점의 크기를 인원수에 따라서 다르게 표현

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='smoker', size='size')
plt.show()
```

# 카테고리 그래프: `catplot()`

* 카테고리 별로 subolot들을 생성하고 통계값(평균, 중앙값, ...) 계산해서, 각각의 subplot에 같은 종류의 그래프들을 시각화.

```python
# 요일별, 흡연여부별 팁의 평균을 막대 그래프로 표현
sns.barplot(data=tips, x='day', y='tip', hue='smoker')
plt.show()
```

```python
sns.catplot(data=tips, x='day', y='tip', col='smoker', kind='bar')
plt.show()
```

* 요일별 팁의 평균 막대 그래프
* 시간(Lunch, Dinner) 별로 subplot을 생성.

```python
sns.catplot(data=tips, x='day', y='tip', hue='sex', col='time', kind='bar')
plt.show()
```

* 요일별 팁의 평균 막대 그래프
* 흡연여부 별로 subplot을 생성
* 시간 별로 subplot을 생성

```python
sns.catplot(data=tips, x='day', y='tip', row='smoker', col='time', kind='bar')
plt.show()
```

* 요일별 성별 팁의 막대 그래프
* 시간 별로 subplot을 생성

```python
sns.catplot(data=tips, x='sex', y='tip', hue='day', col='time', kind='bar')
plt.show()
```

```python
sns.catplot(data=tips, x='day', y='tip', hue='sex', col='time', kind='bar')
plt.show()
```

* 성별 팁의 box plot
* 요일별로 subplot

```python
sns.catplot(data=tips, x='sex', y='tip', col='day', kind='box')
plt.show()
```
